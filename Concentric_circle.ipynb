```python
## Concentric Circles Dataset

import torch 
import numpy as np

def make_circles(n_points=1000, noise=0.07, radius_inner=0.5, radius_outer=1.0):
   n = n_points // 2
   theta = torch.linspace(0, 2 * np.pi, n)

   x0 = torch.stack([
       radius_inner * torch.cos(theta),
       radius_inner * torch.sin(theta)
   ], dim=1)
   x0 += torch.randn_like(x0) * noise

   x1 = torch.stack([
       radius_outer * torch.cos(theta),
       radius_outer * torch.sin(theta)
   ], dim=1)
   x1 += torch.randn_like(x1) * noise

   X = torch.cat([x0, x1], dim=0)
   y = torch.cat([torch.zeros(n), torch.ones(n)]).long()

   return X, y


X, y = make_circles()
print(X.shape, y.shape)

```

```python
import matplotlib.pyplot as plt

def plot_dataset(X, y, title="Dataset"):
   # Convert tensors to numpy for plotting
   X_np = X.detach().cpu().numpy()
   y_np = y.detach().cpu().numpy()

   plt.figure(figsize=(6, 6))
   plt.scatter(X_np[:, 0], X_np[:, 1], c=y_np, s=10)
   plt.title(title)
   plt.xlabel("X1")
   plt.ylabel("X2")
   plt.grid(True)
   plt.show()


# Example usage
X, y = make_circles()
plot_dataset(X, y, title="Concentric Circles")
```
