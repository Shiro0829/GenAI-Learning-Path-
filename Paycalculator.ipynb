```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X_train= np.array([10, 20, 30, 40, 0], dtype=float)
Y_train= np.array([350, 550, 750, 950, 150], dtype=float)

weight = np.random.randn()
bias = np.random.randn()
learning_rate = 0.001

for epoch in range(2000):
    for x, target in zip(X_train, Y_train):
        #Forward pass: Making a prediction 
        prediction = (weight * x) + bias

        error = prediction - target
        weight_gradient = error * x
        bias_gradient = error * 1

        weight = weight - (learning_rate * weight_gradient)
        bias = bias - (learning_rate * bias_gradient)

    if (epoch +1)%400 == 0:
        prediction = (X_train * weight) + bias
        mse_loss = np.mean((prediction - Y_train) ** 2)
        print(f"Epoch {epoch + 1:4d} | Average Loss: {mse_loss:8.2f} | Current hourly Rate (W): {weight:.2f} | Current Base Pay (B): {bias:.2f}") 
#Testing the trained Neuron
print("\n ---Training complete---")

plt.figure(figsize=(8, 5))

plt.scatter(X_train, Y_train, color='blue', label = 'Real Employee Data', zorder=5)

x_line= np.linspace(0,40, 100)
y_line = (weight * x_line) + bias

plt.plot(x_line, y_line, color='red', linewidth=2, label=f'AI Formulae: Line= ({weight:.1f} * Hours) + {bias:.1f}')
plt.title('Employee Pay vs Hours Worked')
plt.xlabel('Hours Worked', fontsize=12)
plt.ylabel('Total Weekly Pay ($)', fontsize=12)
plt.xlim(-2,45)
plt.ylim(0, 1100)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()

print(f"Final learned formulae:Total_Pay= ( hour * {weight:.2f} + {bias:.2f}")

test_hours = 25.0
predicted_pay = (test_hours * weight) + bias
print(f"Predicted pay for working {test_hours} hours: ${predicted_pay:.2f}")
```
