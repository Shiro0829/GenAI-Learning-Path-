```python
import torch
import torch.nn as nn
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# 1. Define the network structure explicitly in this cell
class SimpleNeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size) 
        self.relu = nn.ReLU()                          
        self.fc2 = nn.Linear(hidden_size, output_size)  
    def forward(self, x):
        return self.fc2(self.relu(self.fc1(x)))

print("Step 1: Network defined successfully.")

# 2. Load and scale the data
iris = load_iris()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris.data)

X_tensor = torch.tensor(X_scaled, dtype=torch.float32)
y_tensor = torch.tensor(iris.target, dtype=torch.long)
print(f"Step 2: Loaded dataset with shape {X_tensor.shape}")

# 3. Initialize model (4 inputs -> 16 hidden -> 3 outputs)
iris_model = SimpleNeuralNet(input_size=4, hidden_size=16, output_size=3)
print("Step 3: Model initialized.")

# 4. Test a single forward pass
sample_output = iris_model(X_tensor[:5])

# 5. FORCE OUTPUT VISUALIZATION
print("\n=== FINAL OUTPUT RESULTS ===")
print(sample_output)
print("============================")
```
